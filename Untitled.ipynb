Import neccessary libraries

```python
#data analysis

import pandas as pd 
import numpy as np

#data visualisation

import matplotlib.pyplot as plt
import seaborn as sns

#display all coloumns in data frame and all content of the output

pd.set_option('display.max_columns',None)
pd.set_option('display.max_colwidth', None)
```

```python
Import data
```

```python
df = pd.read_csv('./developer_survey_2020/survey_results_public.csv')
df.head(2)
```

```python
df.shape
```

Identify

Identify the role as Data Scientist or Machine learning specialists or Data or business analyst. And make a new Data Frame for the categories selected.

```python
ds = df[df['DevType']== ('Data scientist or machine learning specialist' or 'Data or business analyst')]
ds.head(2)
```

```python
ds.Ethnicity.unique()
```

```python
pd.DataFrame(ds['Ethnicity'].dropna().str.split(';').to_list()).stack().value_counts()
```

```python
#make a new dataframe for people of colour

dsc = ds[ds['Ethnicity'] != 'White or of European descent']

dsc.head(2)
```

```python
dsc.shape
```

Geographic Analysis:

1.Countries with the highest number of Data Scientists, Machine learning specialists, Business Analysts or Data Analysts?


```python
#listing all the countris
country = dsc.Country.value_counts()

#getting country names and calculating percentages
x = country.index
y = country/country.sum() *100

#create and plot graph
f,ax1 = plt.subplots(figsize=(15,5))
sns.barplot(x=x, y=y, ax=ax1);
ax1.set(xlabel='Countries', ylabel='Percentages %', title="Countries with data scientists, machine learning specialits, data analysts and business analysts of color");
```

```python
#top 10 countries #chose to do top 10 because last dataset was too big
country = dsc.Country.value_counts().head(10)

#getting country names and calculating percentages
x = country.index
y = country/country.sum() *100

#create and plot graph
f,ax1 = plt.subplots(figsize=(15,5))
sns.barplot(x=x, y=y, ax=ax1);
ax1.set(xlabel='Countries', ylabel='Percentages %', title=" Top countries with data scientists, machine learning specialits, data analysts and business analysts of color");
```

In the image above you can see that india hosts the majority of data scientists, machine learning specialits, data analysts and business analysts of people color, with the United States coming in close second. The rest of the countries make up a small percentage of the total populus with most being well below 10 %.

Demographic Analysis:

2.What is the age of most Data Scientists, Machine learning specialists,Business Analysts or Data Analysts?

```python
#creating dataset to graph
age = pd.DataFrame(dsc['Age'].dropna().astype(int).to_list()).stack()
age = age.value_counts()

#getting ages and perecentages of ages
x=age.index
y=age/age.sum()*100

#create and plot graph
f,ax1=plt.subplots(figsize=(16,5))
ax1.set(xlabel='Age', ylabel='Percentage', title='Age of the Data Scientists, Machine learning,specialists,Business Analysts or Data Analysts of colour')
sns.barplot(x=x, y=y, ax=ax1)
plt.show()
```

From the data gathered we can see that the majority of the population in these fields are between 30 and 31 years old, with the largest age group being 26 year olds.

3.Highest Level of Formal Education for Data Scientists, Machine learning specialists, Business Analysts or Data Analysts?

```python
#create dataset to graph
formal_education = pd.DataFrame(ds['EdLevel'].dropna().str.strip().tolist()).stack()
formal_education= formal_education.value_counts()

#getting levels of education and percentages
y = formal_education.index
x = formal_education/formal_education.sum()*100

#create and plot graph
f,ax1 = plt.subplots(figsize=(5,10))
ax1.set(xlabel = 'Percentage %', ylabel = 'Formal Education', 
        title = "Highest level of education for Data Scientists, Machine learning specialists, Business Analysts or Data Analysts of colour")
sns.barplot(x=x, y=y, ax=ax1);
```

The majority of the populus in the field have a masters degree and following in close second are people with a Bachelors.

The Industrial Analysis:

4.How many years of experience do Data Scientists, Machine learning specialists,Business Analysts or Data Analysts have?

```python
#Need to replace some entries in the "YearsCode" column due to them being strings.
years_coding = dsc['YearsCode'].replace({'Less than 1 year': '<1', 'More than 50 years': '>50'}).value_counts()

#getting years of coding and percentages of those years
x = years_coding.index
y = years_coding/years_coding.sum() *100

#create and plot graph
f,ax1 = plt.subplots(figsize=(15,5))
sns.barplot(x=x, y=y, ax=ax1)
ax1.set(xlabel = 'Coding experience in years', ylabel = 'Percentage %', 
        title = "Overall coding experience as a Data Scientists, Machine learning specialists,Business Analysts or Data Analysts of color");
```

From the image above we can see that majority of the populus has less than 10 years of experience coding.
